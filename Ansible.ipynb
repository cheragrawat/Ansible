```python
%%writefile basic_setup.yml
---
- name: Basic configuration script
  hosts: local
  connection: local
  become: false

  tasks:
    - name: Create a folder using Ansible
      file:
        path: /content/my_ansible_folder
        state: directory

    - name: Create a configuration file
      copy:
        dest: /content/my_ansible_folder/config.txt
        content: |
          This file was created by Ansible on Google Colab.
          Ansible is successfully configured and running.
```

```python
!ansible-playbook -i inventory.ini basic_setup.yml
```
